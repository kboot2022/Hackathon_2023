```python
import pandas as da
import numpy as nu
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Table loading, cleaning, and processing
bs=da.read_csv('Dauphin_Island_T_S_O2.csv')
#bs.head()
bs2020=bs[bs.YYYY==2020]
bs2020=bs2020.drop('ss', axis=1)
year=2020
months=nu.arange(1,13,1)
days=nu.arange(1,32,1)
avg=da.DataFrame(columns= ['YYYY', 'MM', 'DD', 'T', 'S', 'O'])
for month in months:
  for day in days:
      filter=(bs2020.MM==month)*(bs2020.DD==day)
      slice=bs2020[filter]
      if slice.empty==True:
        avg.loc[len(avg.index)]= [year, month, day, nu.nan, nu.nan, nu.nan]
      else:
        tempm=slice['T (^oC)'].mean()
        salm= slice['S (PSU)'].mean()
        oxym= slice['O2 (mg/L)'].mean()
        avg.loc[len(avg.index)]= [year, month, day, round(tempm, 4), round(salm,4), round(oxym,4)]
avg.head()
avg.to_csv('Dauphin_Island_2020_Avg.csv')


#bs2020.head()
#bs2020['datestr']=bs2020.iloc[:,0:5].astype(str).apply('-'.join,axis=1)
#bs2020['datestr']=da.to_datetime(bs2020['datestr'],format='%Y-%m-%d-%H-%M')
#bs2020=bs2020.drop(['YYYY','MM', 'DD','hh','mm'], axis=1)



```

```python
#Table loading, cleaning, and processing
bs=da.read_csv('Dauphin_Island_T_S_O2.csv')
#bs.head()
bs=bs.drop(['mm','ss'], axis=1)
months=nu.arange(1,13,1)
years=nu.arange(2010,2023,1)
avg=da.DataFrame(columns= ['YYYY', 'MM', 'T', 'S', 'O'])
for year in years:
  for month in months:
      filter=(bs.MM==month)*(bs.YYYY==year)
      slice=bs[filter]
      if slice.empty==True:
        avg.loc[len(avg.index)]= [year, month, nu.nan, nu.nan, nu.nan]
      else:
        tempm=slice['T (^oC)'].mean()
        salm= slice['S (PSU)'].mean()
        oxym= slice['O2 (mg/L)'].mean()
        avg.loc[len(avg.index)]= [year, month, round(tempm, 4), round(salm,4), round(oxym,4)]
avg.head()
avg.to_csv('Dauphin_Island_2020_MAvg.csv')
```
