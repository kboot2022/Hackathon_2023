```python
import pandas as da
import numpy as nu
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Table loading, cleaning, and processing
#Daily average collector for set year
bs=da.read_csv('River_Q.csv')
#bs.head()
bs2020=bs[bs.YYYY==2020]
bs2020=bs2020.drop('ss', axis=1)
year=2020
months=nu.arange(1,13,1)
days=nu.arange(1,32,1)
avg=da.DataFrame(columns= ['YYYY', 'MM', 'DD', 'T', 'S', 'O'])
for month in months:
  for day in days:
      filter=(bs2020.MM==month)*(bs2020.DD==day)
      slice=bs2020[filter]
      if slice.empty==True:
        avg.loc[len(avg.index)]= [year, month, day, nu.nan, nu.nan, nu.nan]
      else:
        tempm=slice['T (^oC)'].mean()
        salm= slice['S (PSU)'].mean()
        oxym= slice['O2 (mg/L)'].mean()
        avg.loc[len(avg.index)]= [year, month, day, round(tempm, 4), round(salm,4), round(oxym,4)]
avg.head()
avg.to_csv('Dauphin_Island_2020_Avg.csv')


#bs2020.head()
#bs2020['datestr']=bs2020.iloc[:,0:5].astype(str).apply('-'.join,axis=1)
#bs2020['datestr']=da.to_datetime(bs2020['datestr'],format='%Y-%m-%d-%H-%M')
#bs2020=bs2020.drop(['YYYY','MM', 'DD','hh','mm'], axis=1)



```

```python
#Table loading, cleaning, and processing
#Monthly average collector for set year range
bs=da.read_csv('Dauphin_Island_T_S_O2.csv')
#bs.head()
bs=bs.drop(['mm','ss'], axis=1)
months=nu.arange(1,13,1)
years=nu.arange(2003,2023,1)
avg=da.DataFrame(columns= ['YYYY', 'MM', 'T', 'S', 'O'])
for year in years:
  for month in months:
      filter=(bs.MM==month)*(bs.YYYY==year)
      slice=bs[filter]
      if slice.empty==True:
        avg.loc[len(avg.index)]= [year, month, nu.nan, nu.nan, nu.nan]
      else:
        tempm=slice['T (^oC)'].mean()
        salm= slice['S (PSU)'].mean()
        oxym= slice['O2 (mg/L)'].mean()
        avg.loc[len(avg.index)]= [year, month, round(tempm, 4), round(salm,4), round(oxym,4)]
avg.head()
avg.to_csv('Dauphin_Island_2003_2022_MAvg.csv')
```

```python
#Table loading, cleaning, and processing
#Monthly average collector for set year range
bs=da.read_csv('River_Q.csv')
#bs.head()
bs=bs.drop(['mm','ss'], axis=1)
months=nu.arange(1,13,1)
years=nu.arange(2010,2021,1)
avg=da.DataFrame(columns= ['YYYY', 'MM', 'Q'])
for year in years:
  for month in months:
      filter=(bs.MM==month)*(bs.YYYY==year)
      slice=bs[filter]
      if slice.empty==True:
        avg.loc[len(avg.index)]= [year, month, nu.nan]
      else:
        Qm=slice['Q_AR_TR_sum (m^3/s)'].mean()
        avg.loc[len(avg.index)]= [year, month, round(Qm, 4)]
avg.head()
avg.to_csv('QRiver_2010_2020_MAvg.csv')
```

```python
#Concat tables
T1=da.read_csv('Dauphin_Island_2020_MAvg.csv')
T2=da.read_csv('QRiver_2010_2020_MAvg.csv')
T3=da.read_csv('radiation_and_atmospheric.csv')
Variables=da.concat([T1,T2,T3], axis=1)
Variables.to_csv('EverythingButTide.csv')
Variables.head()
```

```python
#Concat wl tables
t1=da.read_csv('wl_8735180_2010_2015.csv')
t2=da.read_csv('wl_8735180_2016_2022.csv')
wl=da.concat([t1,t2], axis=0)
wl=wl.drop(['MM','SS', 'HH'], axis=1)
months=nu.arange(1,13,1)
years=nu.arange(2010,2021,1)
avg=da.DataFrame(columns= ['YYYY', 'MM', 'WL'])
for year in years:
  for month in months:
      filter=(wl.Month==month)*(wl.Year==year)
      slice=wl[filter]
      if slice.empty==True:
        avg.loc[len(avg.index)]= [year, month, nu.nan]
      else:
        Qm=slice['WL'].mean()
        avg.loc[len(avg.index)]= [year, month, round(Qm, 4)]
avg.head()
avg.to_csv('MAvgWL_2010_2020.csv')
t2.head()
#wl.tail()
#Variables.to_csv('EverythingButTide.csv')
```
