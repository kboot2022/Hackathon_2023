```python
%xmode plain
import warnings
warnings.filterwarnings("ignore")
import xarray as xr
import numpy as np
import pandas as pd


### load data 
da1=xr.open_dataset('monthly_radiation.nc',decode_times=True)
da2=xr.open_dataset('wind_airtemp_2003_2022.nc',decode_times=True)
da2
```

```python
## selecting the lat and lon 
# ----------------------------
# vars=[slhf,ssr,sshf]
# for i in range(len(vars)):
#     print('this is'+ vars[i])
#     vars[i]=da1.vars[i].sel(latitude=30.25,longitude=-88.25,method='nearest')
## ---- shld work ..something is messing here.. so I avoided looping . 

slhf=da1.slhf.sel(latitude=30.25,longitude=-88,method='nearest')
ssr=da1.ssr.sel(latitude=30.25,longitude=-88,method='nearest')
sshf=da1.sshf.sel(latitude=30.25,longitude=-88,method='nearest')
str1=da1.str.sel(latitude=30.25,longitude=-88,method='nearest')
u10=da2.u10.sel(latitude=30.25,longitude=-88,time=slice('2010-01-01', '2020-12-01'))

v10=da2.v10.sel(latitude=30.25,longitude=-88,time=slice('2010-01-01', '2020-12-01'))
t2m=da2.t2m.sel(latitude=30.25,longitude=-88,time=slice('2010-01-01', '2020-12-01'))

```

```python
## saving to csv 


#qq=slhf.to_dataframe().to_csv('radiation.csv')

slhf_1=slhf.to_dataframe()
ssr_1=ssr.to_dataframe()
sshf_1=sshf.to_dataframe()
str_1=str1.to_dataframe()
u10=u10.to_dataframe()
v10=v10.to_dataframe()
t2m=t2m.to_dataframe()
result = pd.concat([slhf_1, ssr_1['ssr'],sshf_1['sshf'],str_1['str'],u10['u10'],v10['v10'],t2m['t2m']], axis=1)
result
```
