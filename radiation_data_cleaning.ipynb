```python
%xmode plain
import warnings
warnings.filterwarnings("ignore")
import xarray as xr
import numpy as np
import pandas as pd


### load data 
da1=xr.open_dataset('adaptor.mars.internal-1686251817.1035593-11923-2-20c5e3cc-81ac-4d64-8618-a3d86a646f80.nc',decode_times=True)


## selecting the lat and lon 
# ----------------------------
# vars=[slhf,ssr,sshf]
# for i in range(len(vars)):
#     print('this is'+ vars[i])
#     vars[i]=da1.vars[i].sel(latitude=30.25,longitude=-88.25,method='nearest')
## ---- shld work ..something is messing here.. so I avoided looping . 

slhf=da1.slhf.sel(latitude=30.25,longitude=-88.25,method='nearest')
ssr=da1.ssr.sel(latitude=30.25,longitude=-88.25,method='nearest')
sshf=da1.sshf.sel(latitude=30.25,longitude=-88.25,method='nearest')
sshf.time
```

```python
### finding the daily mean from the hourly data

slhf=slhf.groupby("time.day")#.mean(dim='time')
ssr=ssr.groupby("time.day").mean(dim='time')
sshf=sshf.groupby("time.day").mean(dim='time')
slhf
```

```python
slhf=slhf.resample("D").mean(dim="time")
```

```python
## saving to csv 


#qq=slhf.to_dataframe().to_csv('radiation.csv')

slhf_1=slhf.to_dataframe()
ssr_1=ssr.to_dataframe()
sshf_1=sshf.to_dataframe()

result = pd.concat([slhf_1, ssr_1['ssr'],sshf_1['sshf']], axis=1)
result
```

```python
ssr.to_dataframe().to_csv('ssr.csv')
sshf.to_dataframe().to_csv('sshf.csv')
```
